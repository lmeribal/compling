## 4. Pymorphy vs Mystem (2 балла)

Лемматизируйте очищенный текст с помощью mystem и pymorphy. Сравните результаты: есть ли разница, что лучше. 

Для пайморфи используйте токенизацию из mystem, чтобы исключить влияние токенизации на результат.

```python
from pymystem3 import Mystem
from pymorphy2 import MorphAnalyzer

mystem = Mystem()
pymorphy = MorphAnalyzer()
```

```python
%%time
lemmas_mystem = "".join(mystem.lemmatize(text))
```

```python
%%time
pymorphy_lemmatized = []
analyzed = mystem.analyze(text)

for token in analyzed:
    parsed = pymorphy.parse(token['text'])[0]
    pymorphy_lemmatized.append(parsed.normal_form)
lemmas_pymorphy = "".join(pymorphy_lemmatized)
```

Разница между леммами:

```python
for token_1, token_2 in zip(lemmas_mystem.split(), lemmas_pymorphy.split()):
    if token_1 != token_2:
        print(token_1, token_2)
```

**Выводы:**

1. Стандартный pymorphy2 в приведённом случае работает медленнее MyStem
2. MyStem игнорирует букву "ё", корректно приводит глаголы к инфинитиву
3. Pymoprhy приводит "т.к." к нелогичному виду "так.к.", распознает глагол в словах "юзабилити", "врядли", и приводит их к инфинитиву. Иногда Pymorphy возвращает форму совершенного вида вместо несовершенного (собрать, повысить, отследить...). Слово "мой" приведено к форме "мыть", хотя в контексте обозначает местоимение. Тем не менее, в отличии от MyStem, pymorphy верно отнес токен "стоит" к начальной форме "стоить", а MyStem – к форме "стоять"

По скорости, возможностям и результату я бы отдал предпочтение MyStem – лемматизированный текст получился корректнее и проанализирован быстрее.

С помощью Mystem разделите текст из первого задания (очищенный) на предложения. Сравните разбиение с razdel.sentenize

**MyStem**

```python
%%time

!./mystem -isc --format json text.txt text_parsed.txt

import json
t = [json.loads(line) for line in open('text_parsed.txt')]

result = []

for i in range(len(t)):
    tmp_lst = []
    for d in t[i]:
        tmp_lst.append(d['text'])
    tmp_lst = "".join(tmp_lst).split("\\s")
    for s in tmp_lst:
        if s != '\n':
            result.append(s.replace('\n', ''))
        
print(f"Количество предложений: {len(result)}")
for el in result:
    print(el)
```

```python
%%time

from razdel import sentenize

result = [s.text for s in sentenize(text)]

print(f"Количество предложений: {len(result)}")
for el in result:
    print(el)
```

## 6*. Дизамбигуация в Mystem (2 балла)

*необязательное задание на 10 баллов

Одна из фичей майстема - контекстная дизамбигуация (т.е. определение правильного разбора в зависимости от контекста).

Запустите mystem на очищенном тексте с дизамбигуацией и без и проверьте есть ли какие-то различия.

```python
mystem_disamb = Mystem(disambiguation=True)
mystem_normal = Mystem(disambiguation=False)

lemmas = mystem_normal.lemmatize(text)
lemmas_dis = mystem_disamb.lemmatize(text)
''.join(lemmas_dis) == ''.join(lemmas)
```

```python
for token_1, token_2 in zip(lemmas, lemmas_dis):
    if token_1 != token_2:
        print(f"Без дизамбигуации: {token_1}\nС дизамбигуацией: {token_2}\n")
```

Различия есть: например, самое важное в словах "стоять" и "стоить". Контекст – предложение "Стоит почти 100 долларов, имеет возможность отследить 250 000 посещений на ста страницах". MyStem с дизамбигуацией верно относит "стоит" к глаголу с начальной форме "стоить" и обозначающим ценность. Версия без дизамбигуации относит глагол к неверной форме.
